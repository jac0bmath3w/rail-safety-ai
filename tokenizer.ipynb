```python
from transformers import AutoTokenizer

# 1. Load the Llama 3.2 Tokenizer
model_id = "unsloth/Llama-3.2-3B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id)

# 2. A complex railroad sentence from your manuals
sentence = "Positive Train Control (PTC) is mandatory for hazardous material compliance."

# 3. Tokenize!
tokens = tokenizer.tokenize(sentence)
ids = tokenizer.encode(sentence)

print(f"Original: {sentence}")
print(f"Tokens: {tokens}")
print(f"Token IDs: {ids}")

# 4. Challenge: Decode just one ID to see the sub-word
print(f"Token ID {ids[1]} is: {tokenizer.decode([ids[1]])}")
```
